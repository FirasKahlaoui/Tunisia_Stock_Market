# Check the DataFrame
*-  if closingPrice < lowestPrice then delete the row*<br>
*-  if openingPrice > highestPrice then delete the row*<br>
*-  if closingPrice > highestPrice then delete the row*<br>
*-  if openingPrice < lowestPrice then delete the row*<br>
*-  if lowestPrice > highestPrice then delete the row*<br>
*- no negative values*

```python
invalid_rows = df[(df['closingPrice'] < df['lowestPrice']) | 
                   (df['openingPrice'] > df['highestPrice']) |
                   (df['closingPrice'] > df['highestPrice']) |
                   (df['openingPrice'] < df['lowestPrice']) |
                   (df['lowestPrice'] > df['highestPrice']) |
                   (df['openingPrice'] < 0) |
                   (df['highestPrice'] < 0) |
                   (df['lowestPrice'] < 0) |
                   (df['closingPrice'] < 0) |
                    (df['volume'] < 0) ]


print(invalid_rows)
```

```python
# Delete the invalid rows from the dataframe
df = df[~df.index.isin(invalid_rows.index)]
df
```

` => The DataFrame is checked for the above conditions and the rows are deleted accordingly.`

```python
# Export the cleaned data to a new csv file
df.to_csv('./data/checked_weekly_stock_market.csv', index=False)
```
